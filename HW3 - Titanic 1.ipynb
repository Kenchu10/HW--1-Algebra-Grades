```python
import pandas as pd
import numpy as np


Location=r"C:\Users\Albert\datasets\titanic3.csv"
df=pd.read_csv(Location)
df.head()
```

```python
df.tail()
```

```python
# Show records with age less than 1

df.loc[df['age'] < 1].head()
```

```python
# If the age field is less than 1, set it to 1

df.loc[(df['age'] < 1, 'age')] = 1
```

```python
# Make sure it worked

df.loc[df['age'] < 1]
```

```python
# Find the mean age of all the passengers on the ship

df["age"].mean()
```

```python
# Fill the age empty field with the mean age of 29

df["age"].fillna(int(df["age"].mean()), inplace=True)
```

```python
# Replacement survived from 1/0 to Yes/No

df['survived_YN'] = np.where(df['survived'] < 1, 'No', 'Yes')
```

```python
# Write out file with fixed records

df2=pd.DataFrame(data=df)
df2.to_excel('titanic3.xlsx',index=False)
```
