```python
import pandas as pd
import numpy as np

Location=r"C:\Users\Albert\datasets\axisdata.xlsx"
df=pd.read_excel(Location, names=['Fname','Lname','Gender','Hours Worked','SalesTraining','Years Experience','Cars Sold'])

df['All Emp'] = "EMP"
df.head()
```

```python
print "Average cars sold per month"

pd2=pd.pivot_table(df,index=['All Emp'], aggfunc='mean', values=["Cars Sold"])
pd2
```

```python
print "Max cars sold per month"

pd2=pd.pivot_table(df,index=['All Emp'], aggfunc='max', values=["Cars Sold"])


pd2
```

```python
print "Min cars sold per month"

pd2=pd.pivot_table(df, index=['All Emp'], aggfunc='min', values=["Cars Sold"], margins=True)
pd2
```

```python
print "Average cars sold per month by gender:"

pd2=pd.pivot_table(df, index=['Gender'], aggfunc='mean', values=["Cars Sold"], margins=True)
pd2
```

```python
soldmorethan3cars=df['Cars Sold'] > 3

df[soldmorethan3cars].tail()
```

```python
print "Average hour worked per month by people sold more than 3 cars:"

pd2=pd.pivot_table(df[soldmorethan3cars], index=['All Emp'], aggfunc='mean', values=["Hours Worked"])

pd2
```

```python
print "Average years of experience:"

pd2=pd.pivot_table(df, index=['All Emp'], aggfunc='mean', values=["Years Experience"])

pd2
```

```python
print "Average year of experience by people sold more than 3 cars per month:"

pd2=pd.pivot_table(df[soldmorethan3cars], index=['All Emp'], aggfunc='mean', values=["Years Experience"])

pd2
```

```python
print "Average cars sole per month by people had Sales Training:"

pd2=pd.pivot_table(df, index=['SalesTraining'], aggfunc='mean', values=["Cars Sold"])

pd2
```
