```python
Location="C:\Users\Albert\datasets\grades_all.xlsx"
df=pd.read_excel(Location,skiprow=1)
df.head()
```

```python
df.tail()
```

```python
passing=df['Grade'] >= 70
failed =df['Grade'] <  70
female=df['Gender'] == "female"
```

```python
print "Total students in the System:"

df2=pd.pivot_table(df,index=['Gender','Year'], aggfunc='count', values=["Address"], margins=True)

df2.columns = ['# of people']
df2
```

```python
print "Number of students failed:"

df2=pd.pivot_table(df[failed],index=['Gender','Year'], aggfunc='count', values=['Address'], margins=True)

df2.columns = ['# of people']
df2
```

```python
print "Number of students passing:"

df2=pd.pivot_table(df[passing],index=['Gender','Year'], aggfunc='count', values=['Address'], margins=True)

df2.columns = ['# of people']
df2
```

```python
print "Averages of grades and hours of study for all students:"

pd.pivot_table(df,index=['Gender','Year'], aggfunc='mean', values=['Grade','Hours'], margins=True)
```

```python
print "Averages of grades and hours of study for all student with a passing grade:"

pd.pivot_table(df[passing],index=['Gender','Year'], aggfunc='mean', values=['Grade','Hours'], margins=True)
```

```python
print "Averages of grades and hours of study for all female student with a passing grade:"

pd.pivot_table(df[female & passing],index=['Gender','Year'], aggfunc='mean', values=['Grade','Hours'], margins=True)
```
